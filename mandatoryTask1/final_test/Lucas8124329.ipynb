## 1. Programa para sobrepor quadriculado em imagem de cinza

Segue-se a função criada para sobrepor um quadriculado em uma imagem de cinza.

```python
def program1(f, delta):
    f1 = np.array(f, copy=True) # copia profunda
    
    f1[::delta,:] = 255 # linhas horizontais
    f1[:,::delta] = 255 # linhas verticais

    return f1
```

Faremos uma implementação da função criada. Dados:

```python
# Dados:
delta = 10 # por exemplo, se delta = 10
f = mpimg.imread('data/retina.tif')
plt.imshow(f, cmap='gray')
```

Aplicaremos a função:

```python
# Aplicacao da funcao
f1 = program1(f, delta)
```

Verificaremos a imagem resultante:

```python
# Faremos um zoom na imagem
plt.imshow(f1[:150,:150],cmap='gray')
```

## 2. Programa de imagens de quadrados cinzas

Temos que `program2` gera a imagem com os quadrados cinzas.

```python
def program2 (isImg):
    if (isImg):
        H,W = 300,600
    else:
        H,W = 6,12

    # quadrado interior
    h1,h2 = int(H/3), int(2*H/3)    # altura e comprimento
    w1,w2 = int(2*W/3), int(5*W/6)  # posicao inicio e fim na horizontal do quad.

    f2 = np.zeros( (H,W), 'uint8')  # gerando matriz
    f2[:,:H] = 64      # gerando quad. esquerdo
    f2[:,H:] = 192     # gerando quad. direito
    
    f2[h1:h2, h1:h2] = 128 # gerando quad. central esquerdo
    f2[h1:h2, w1:w2] = 128 # gerando quad. central direito

    plt.imshow(f2,cmap='gray')
    
    return f2
```

A seguir, teremos um exemplo de implementação quando `isImg` for verdadeira.

Peculiarmente, a imagem desenhada pela função `plt.imshow()` seta o maior tom de cinza disponível na imagem gerada e escala ela de 0-255. Portanto o cinza 192 é considerada como 255.

Note que no `colorbar`, o nível de tom de cinza vai de 64 a 192.

```python
isImg = True
program2 (isImg)
plt.colorbar()
```

Mas vejamos quando `isImg` é falsa. A imagem gerada é gerada da mesma forma segundo a anterior, mas analisando a matriz gerada, vejamos que a saída está correta.

```python
isImg = False
program2 (isImg)
plt.colorbar()    
```

## Aplicando bordas

No entando, aplicando uma borda na imagem resulta na visualização esperada. Usaremos uma borda com 1 pixel de espessura.

Note que o `colorbar` varia de 0 a 255:

```python
isImg = True
f_new = program2 (isImg)

f_new[   :1,   :  ] = 0   # frame superior
f_new[-1:  ,   :  ] = 0 # frame inferior
f_new[   : ,   :1]  = 255 # frame esquerdo
f_new[   : , -1:  ] = 255 # frame direito

plt.imshow(f_new,cmap='gray')
plt.colorbar()
```

## 3. Programa para rotacionar uma imagem de 90 graus no sentido horário

Temos em seguinte a função `program3`, que rotaciona 90 graus no sentido horário.

```python
def program3 (f):
    g = f.transpose() # calculo da transposta
    g = g[:,::-1]     # refletimos na horizontal
    plt.imshow(g, cmap='gray')
```

Utilizaremos a seguinte imagem como exemplo de aplicação:

```python
f = mpimg.imread('data/cameraman.tif')
plt.imshow(f, cmap='gray')
```

Resultado:

```python
program3(f)
```
